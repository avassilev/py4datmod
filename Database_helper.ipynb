```python
import pandas as pd

firms = pd.DataFrame({
    "firm_id": [1, 2],
    "firm_name": ["Alpha Corp", "Beta Ltd"],
    "sector": ["Manufacturing", "Retail"]
})

products = pd.DataFrame({
    "product_id": [101, 102, 201],
    "firm_id": [1, 1, 2],
    "product_name": ["Widget", "Gadget", "Thing"],
    "price": [10.0, 15.0, 8.0]
})

sales = pd.DataFrame({
    "sale_id": range(1, 7),
    "product_id": [101, 101, 102, 201, 201, 201],
    "year": [2021, 2022, 2022, 2021, 2022, 2023],
    "quantity": [100, 120, 80, 200, 180, 210]
})
```

```python
import sqlite3

conn = sqlite3.connect("company_data.db")

firms.to_sql("firms", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)
sales.to_sql("sales", conn, if_exists="replace", index=False)

conn.close()
```
